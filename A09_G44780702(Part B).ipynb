# (iii)

Provide the histogram and density plot of the inter-arrival times for the top five busiest airports in the list provided. All plots should be in the same figure.

```python
# First we need to find 5 busiest airports
dest=flights['dest'].value_counts()
dest.head(5)
```

```python
# We subset the data into five airports and sort the time
# Since arr_time column has many null values ,we use sched_arr_tine to plot.
ORD=flights[flights.dest=="ORD"].sort_values(["year","month","day","sched_arr_time"])
ATL=flights[flights.dest=="ATL"].sort_values(["year","month","day","sched_arr_time"])
LAX=flights[flights.dest=="LAX"].sort_values(["year","month","day","sched_arr_time"])
BOS=flights[flights.dest=="BOS"].sort_values(["year","month","day","sched_arr_time"])
MCO=flights[flights.dest=="MCO"].sort_values(["year","month","day","sched_arr_time"])
```

In order to compute the inter-arrival times, we need to combine year, month and day and arrival time into one column and convert the data type into time

```python
# Convert the data type to string first 
# ORD
ORD['sched_arr_time'] = ORD.sched_arr_time.map(lambda x:str(x).zfill(4))   # shown in HHMM
ORD['year'] = ORD.year.map(lambda x:str(x))
ORD['month'] = ORD.month.map(lambda x:str(x))
ORD['day'] = ORD.day.map(lambda x:str(x))
# ATL
ATL['sched_arr_time'] = ATL.sched_arr_time.map(lambda x:str(x).zfill(4))   # shown in HHMM
ATL['year'] = ATL.year.map(lambda x:str(x))
ATL['month'] = ATL.month.map(lambda x:str(x))
ATL['day'] =ATL.day.map(lambda x:str(x))
# LAX
LAX['sched_arr_time'] = LAX.sched_arr_time.map(lambda x:str(x).zfill(4))   # shown in HHMM
LAX['year'] = LAX.year.map(lambda x:str(x))
LAX['month'] = LAX.month.map(lambda x:str(x))
LAX['day'] = LAX.day.map(lambda x:str(x))
# BOS
BOS['sched_arr_time'] = BOS.sched_arr_time.map(lambda x:str(x).zfill(4))   # shown in HHMM
BOS['year'] = BOS.year.map(lambda x:str(x))
BOS['month'] = BOS.month.map(lambda x:str(x))
BOS['day'] = BOS.day.map(lambda x:str(x))
# MCO
MCO['sched_arr_time'] = MCO.sched_arr_time.map(lambda x:str(x).zfill(4))   # shown in HHMM
MCO['year'] = MCO.year.map(lambda x:str(x))
MCO['month'] = MCO.month.map(lambda x:str(x))
MCO['day'] = MCO.day.map(lambda x:str(x))
```

```python
## We combine these four columns into  one column called ad_arr_time
ORD.insert(loc=0,column="ad_arr_time",value=ORD.apply(lambda x: x['year']+"-"+x['month']+"-"+x['day']+" "+x["sched_arr_time"], axis=1))
ATL.insert(loc=0,column="ad_arr_time",value=ATL.apply(lambda x: x['year']+"-"+x['month']+"-"+x['day']+" "+x["sched_arr_time"], axis=1))
LAX.insert(loc=0,column="ad_arr_time",value=LAX.apply(lambda x: x['year']+"-"+x['month']+"-"+x['day']+" "+x["sched_arr_time"], axis=1))
BOS.insert(loc=0,column="ad_arr_time",value=BOS.apply(lambda x: x['year']+"-"+x['month']+"-"+x['day']+" "+x["sched_arr_time"], axis=1))
MCO.insert(loc=0,column="ad_arr_time",value=MCO.apply(lambda x: x['year']+"-"+x['month']+"-"+x['day']+" "+x["sched_arr_time"], axis=1))
```

```python
## We check one of them to make sure everything goes well.
ATL.head(10)
```

```python
# we import the library to convert the column ad_arr_time into datetime format
import datetime as dt
import numpy as np
```

```python
# convert the string into format time
ORD["ad_arr_time"]=ORD['ad_arr_time'].apply(lambda x: pd.datetime.strptime(x,'%Y-%m-%d %H%M'))
ATL["ad_arr_time"]=ATL['ad_arr_time'].apply(lambda x: pd.datetime.strptime(x,'%Y-%m-%d %H%M'))
LAX["ad_arr_time"]=LAX['ad_arr_time'].apply(lambda x: pd.datetime.strptime(x,'%Y-%m-%d %H%M'))
BOS["ad_arr_time"]=BOS['ad_arr_time'].apply(lambda x: pd.datetime.strptime(x,'%Y-%m-%d %H%M'))
MCO["ad_arr_time"]=MCO['ad_arr_time'].apply(lambda x: pd.datetime.strptime(x,'%Y-%m-%d %H%M'))
```

```python
# we check one of them 
ORD.head(50)
```

Then, we conpute the inter arrival times by "diff()" and convert the time difference into minutes. We store them into a new dataframe called XXX_interarrival

```python
ORD_interarrival=pd.DataFrame(ORD['ad_arr_time'].diff().astype('timedelta64[m]'))
ATL_interarrival=pd.DataFrame(ATL['ad_arr_time'].diff().astype('timedelta64[m]'))
LAX_interarrival=pd.DataFrame(LAX['ad_arr_time'].diff().astype('timedelta64[m]'))
BOS_interarrival=pd.DataFrame(BOS['ad_arr_time'].diff().astype('timedelta64[m]'))
MCO_interarrival=pd.DataFrame(MCO['ad_arr_time'].diff().astype('timedelta64[m]'))
```
